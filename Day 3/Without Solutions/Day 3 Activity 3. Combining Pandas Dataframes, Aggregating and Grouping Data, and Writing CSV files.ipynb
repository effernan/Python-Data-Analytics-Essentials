```python
# Solution 1
import pandas as pd
import numpy as np
```

```python
#Solution 2
data_1 = pd.read_csv("join/dickens_locations.csv", index_col = 0)
data_2 = pd.read_csv("join/dickens_number_words.csv", index_col = 0)
```

```python
# Solution 3
data_3 = pd.merge(data_1, data_2, on = "Titles", how = "inner")
data_3
```

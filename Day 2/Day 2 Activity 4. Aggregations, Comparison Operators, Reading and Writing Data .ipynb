```python
# Solution 1

np.greater(total_characters_UK, total_characters_US) 
```

This output is telling us that Charles Dickens (UK) has more characters in his books than Mark Twain (US), but Edith Wharton (US) has more characters than Elizabeth Gaskell (US). Remeber that we used position 1 for male characters and position 2 for female characters!

```python
# Solution 2
np.greater(np.sum(total_characters_UK), np.sum(total_characters_US))
```

That output is telling us that overall, if we sum Charles Dickens + Elizabeth Gaskell and Mark Twain + Edith Wharton, we are getting a higher number of characters in the UK rather than in the US.

```python
# Solution 3.
np.greater(max_characters_UK, max_characters_US)
```

That output is telling us that Charles Dickens has less max characters than Mark Twain and Elizabeth Gaskell has less max characters than Edith Wharton. So: the US has actually a higher peak of maximum number of characters in a book even if the UK has a total higher number of characters overall

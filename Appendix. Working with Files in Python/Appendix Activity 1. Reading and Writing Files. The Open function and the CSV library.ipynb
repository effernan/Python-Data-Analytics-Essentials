# 2. The CSV Python Library

And now let's practice how to use the csv python library!

   * 1. First import the csv library
   * 2. Now use the Open function combined with the csv.reader function and open the file "Edith_Wharton_dates.csv" that is inside the data folder.
   * 3. Modify row 4 in the data variable (with whatever text you like)
   * 4. Now print that into a csv file. Remember to write a path indicating that you want that inside the data folder.

```python
#Solution 1
import csv
```

```python
#Solution 2

with open("data/Edith_Wharton_dates.csv", "r", encoding = "utf-8") as f:
    reader = csv.reader(f)
    data = list(reader)
    
data
```

```python
#Solution 3

data[3] = ["this is a test"]

data
```

```python
# Solution 4

with open("data/data_modified.csv", "w", encoding="utf-8", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
    
file.close()
```
